```python
# ===============================
# 1. IMPORTS
# ===============================
import pandas as pd
import numpy as np
import os
import joblib

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# ===============================
# 2. LOAD DATA
# ===============================
df = pd.read_csv("india_housing_prices.csv")


# ===============================
# 3. FEATURE ENGINEERING
# ===============================

# Encode categorical columns
le_city = LabelEncoder()
le_type = LabelEncoder()

df["City"] = le_city.fit_transform(df["City"])
df["Property_Type"] = le_type.fit_transform(df["Property_Type"])

# Create new features
df["Price_per_SqFt"] = (df["Price_in_Lakhs"] * 100000) / df["Size_in_SqFt"]
df["Age_of_Property"] = 2026 - df["Year_Built"]

# ===============================
# 4. FEATURES & TARGETS (FIXED 🔥)
# ===============================
features = [
    "City",
    "Property_Type",
    "BHK",
    "Size_in_SqFt",
    "Price_per_SqFt",
    "Age_of_Property"
]

# Create classification target
df["Good_Investment"] = (
    df["Price_in_Lakhs"] < df["Price_in_Lakhs"].median()
).astype(int)

X = df[features]

# ✅ ADD THIS (YOU MISSED THIS)
y_class = df["Good_Investment"]
y_reg = df["Price_in_Lakhs"]

# ===============================
# 5. TRAIN TEST SPLIT (FIXED)
# ===============================
X_train, X_test, y_train, y_test = train_test_split(
    X, y_class, test_size=0.2, random_state=42
)

X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

# ===============================
# 6. TRAIN MODELS
# ===============================
clf = RandomForestClassifier(
    n_estimators=50,
    max_depth=10,
    random_state=42
)

reg = RandomForestRegressor(
    n_estimators=100,
    max_depth=15,
    random_state=42
)

clf.fit(X_train, y_train)
reg.fit(X_train_r, y_train_r)

# ===============================
# 7. SAVE MODELS + ENCODERS
# ===============================
os.makedirs("models", exist_ok=True)

joblib.dump(clf, "models/classification_model.pkl", compress=3)
joblib.dump(reg, "models/regression_model.pkl", compress=3)

# ✅ ALSO SAVE THESE (IMPORTANT FOR STREAMLIT)
joblib.dump(le_city, "models/city_encoder.pkl")
joblib.dump(le_type, "models/type_encoder.pkl")
joblib.dump(features, "models/features.pkl")

print("✅ Models trained and saved successfully!")
```
